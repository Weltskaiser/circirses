```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("resM1P1.csv", sep=";", na_values="NA")
print("Dataframe original :")
print(df.head())
```

```python
df = df.set_index("plant")
df = df.dropna(axis=0)
print("Dataframe remodelé :")
print(df.head())
print(df.describe())
```

```python
plt.scatter(df.loc[:, 'markA.1'], df.loc[:, 'markA.2'])
plt.show()
```

```python
from scipy.spatial.distance import pdist, squareform
dist_1 = pdist(df['markA.1'].values.reshape(-1, 1), 'euclidean')
dist_mat_1 = np.array(squareform(dist_1))
# print(dist_mat_1)
dist_2 = pdist(df['markA.2'].values.reshape(-1, 1), 'euclidean')
dist_mat_2 = np.array(squareform(dist_2))
# print(dist_mat_2)
dist_mat_comb = dist_mat_1 + dist_mat_2
print(dist_mat_comb)
plt.imshow(dist_mat_comb, cmap='hot', interpolation='nearest')
```
