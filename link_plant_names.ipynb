```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_m1p1_locus = pd.read_csv("resM1P1.csv", sep=";", na_values="NA")
# df_m1p1_locus.set_index("plant", inplace=True)
print(df_m1p1_locus)
```

```python
df_plant_names = pd.read_csv("plant_names.csv", sep=";", na_values="NA")
df_plant_names.dropna(axis=0, inplace=True)
print(df_plant_names)
```

```python
#plant_names = df_plant_names['name'].to_list()
df_plant_names.sort_values(by="name", inplace=True)
df_plant_names.reset_index(drop=True, inplace=True)
print(df_plant_names)
```

```python
df = pd.concat([df_m1p1_locus, df_plant_names], axis=1)
print(df)
```

```python
out_path = "resM1P1_plant_names.csv"
df.to_csv(out_path, index=False, na_rep='NA')
```
