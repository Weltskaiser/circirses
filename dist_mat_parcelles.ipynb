```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_m1p1 = pd.read_csv("resM1P1_plant_names.csv", sep=",", na_values="NA")
df_m1p1.dropna(axis=0, inplace=True)
df_m1p1.rename(columns={'markA.1': 'A1', 'markA.2': 'A2'}, inplace=True)
print(df_m1p1)
```

```python
mf01 = df_m1p1[df_m1p1['name'].str.startswith('MF01')]
mf05 = df_m1p1[df_m1p1['name'].str.startswith('MF05')]
mf08 = df_m1p1[df_m1p1['name'].str.startswith('MF08')]
mf09 = df_m1p1[df_m1p1['name'].str.startswith('MF09')]
mf13 = df_m1p1[df_m1p1['name'].str.startswith('MF13')]
mf15 = df_m1p1[df_m1p1['name'].str.startswith('MF15')]
mf17 = df_m1p1[df_m1p1['name'].str.startswith('MF17')]
mf18 = df_m1p1[df_m1p1['name'].str.startswith('MF18')]
```

```python
plt.scatter(mf01['A1'], mf01['A2'], label='MF01 ({} obs)'.format(len(mf01)))
plt.scatter(mf05['A1'], mf05['A2'], label='MF05 ({} obs)'.format(len(mf05)))
plt.scatter(mf08['A1'], mf08['A2'], label='MF08 ({} obs)'.format(len(mf08)))
plt.scatter(mf09['A1'], mf09['A2'], label='MF09 ({} obs)'.format(len(mf09)))
plt.scatter(mf13['A1'], mf13['A2'], label='MF13 ({} obs)'.format(len(mf13)))
plt.scatter(mf15['A1'], mf15['A2'], label='MF15 ({} obs)'.format(len(mf15)))
plt.scatter(mf17['A1'], mf17['A2'], label='MF17 ({} obs)'.format(len(mf17)))
plt.scatter(mf18['A1'], mf18['A2'], label='MF18 ({} obs)'.format(len(mf18)))
plt.legend()
plt.show()
```

```python
from scipy.spatial.distance import pdist, squareform
dist_mat_1 = np.array(squareform(pdist(df_m1p1['A1'].values.reshape(-1, 1), 'euclidean')))
dist_mat_2 = np.array(squareform(pdist(df_m1p1['A2'].values.reshape(-1, 1), 'euclidean')))
dist_mat_comb_m1p1 = dist_mat_1 + dist_mat_2
print(dist_mat_comb_m1p1)
```

```python
plt.imshow(dist_mat_comb_m1p1, cmap='hot')
```
